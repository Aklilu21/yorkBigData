**10) What is the name for the 3 primary measures of central tendency and give their definition** 

1. Mean: it's the average value for a sample or population (sum of all values divided by the number of values).
2. Median: it's the middle value, that is, the value for which there are the same number of values above and below of it.
3. Mode: it's the most common value.

**47) Give a very general and short definition of what an Analytical Model is or is supposed to do?**

An analytical model is a simplified representation of reality that is used to try to answer some questions about a problem, an opportunity or an observation.

**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

In supervised learning training data is "labeled", that is, for every input the corresponding target is provided and after sufficient training the algorithm will be able to provide target for any new input in a "reasonable" way. On the other hand, in unsupervised learning training data is "unlabeled", therefore only a set of inputs is provided, and the algorithm should be able to find the structure or relationships between different inputs.

**49) Briefly describe the differences between "classification" and "clustering".**

- In classification, it is first learned what goes with what (groups are predefined) and then that knowledge is applied to new samples to classify them. Classification is the result of supervised learning. It requires training data.

- In clustering groups or cluster are not predefined (they are provided by the algorithm based on the similarities of data instances to each other). Clustering is the result of unsupervised learning. It does not require training data.

**50) Name 3 different "kinds" of Analytical Models and provide a brief definition of each. **

- __Deterministic models__: In these models, the output of the model is fully determined by the parameter values and the initial conditions.
- __Stochastic models__: these models possess some inherent randomness. The same set of parameter values and initial conditions will lead to an ensemble of different outputs.
- __Algorithm__: is a procedure for solving a problem, based on conducting a sequence of specified actions in order to transform some inputs into outputs.
