**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

In supervised learning, mathematical functions are inferred from labeled training data - i.e. the output datasets are provided.
In unsupervised learning, there is no labeled training data as outputs. Machine learning algorithms and input data must be used to explore the data to find hidden patterns or groupings in the data.

**50) Name 3 different "kinds" of Analytical Models and provide a brief definition of each. **

Three different kinds of analytical models are:

1)DESCRIPTIVE - This is the first level of data processing, which summarizes or condenses big data into smaller useful segments of information, and possibly prepares the data for further analysis.

2)PREDICTIVE - This is the practice of extracting information from data sets in order to determine patterns and predict future outcomes and trends.  It does not, however, tell you what 'will' happen in the future.  It predicts what 'might' happen because it is probabilistic.

3)PRESCRIPTIVE - This relates to the process of finding the best course of action for a given situation, and typically occurs after descriptive and predictive analytics have been processed first.  In addition to predicting future outcomes, prescriptive analytics will suggest actions to benefit from the predictions and show the implications of each decision option that is presented.  Prescriptive analytics can continually take in new data to re-predict and re-prescribe if appropriate.

Big Data differs from Small Data in several ways.  The most prominent ones are:  (1) The sheer volume.  There is no absolute level for big data volumes but it is easily in the range of terabytes or petabytes, so it relies on advanced computer processing technologies to go through large volumes relatively quickly. **EXTRA CREDIT 53) What are WMD's according to Cathy O'Neil?**

WMD's  (Weapons of Math Destruction) are high-powered applications which combine mathematical power and technology to produce outputs which are initially viewed as positive/beneficial/profitable but, upon further analysis, point to model flaws and algorithm flaws which cause bias, human prejudice, misunderstanding and overall unfairness.

An example of a WMD in Cathy O'Neil's book is a teacher assessment tool which was supposed to rank all teachers in Washington, DC such that the worst performers were fired from their jobs.  How well students performed academically was supposed to be mathematically attributed to teacher performance plus the modeling of many other factors.  Some good teachers were given low scores and fired undeservingly.

WMD's end up being destructive because the key decision-makers have blinders on - - they were sold on how great the WMD was by its creators, and its outputs seemed to be credible initially, so not enough attention was paid to the possibilities of error and unfairness and the WMD continued to be used when instead it should have been stopped completely.



**EXTRA CREDIT 54) Why do vicious feedback loops matter in data analytics?**

Vicious feedback loops matter because they are self-reinforcing.  The outputs of a data analytics project are based on its model logic, and we know that 'models are wrong' - i.e. they are only representations of reality.  When data analytics are treated as absolutes, then those who are impacted may be treated unfairly or use resources unwisely. 

An example is the WMD description of US News & World Report's annual ranking of colleges and universities.  75% of each school's ranking comes from a model which was not developed in a proper scientific way.  Someone decided on the tracking of statistics which would point accurately towards the correct ranking of each school. The quality of an education is highly subjective, but this ranking system used data for 75% of the ranking.  Where the vicious feedback loops matter is the reaction of each school to its results.  At the bottom ranks, these schools could not get out from under their low scores to improve because readers of the rankings would reinforce the low rankings by not applying there (students) and not teaching there (professors would leave over time to more prestigious schools). To try to improve rankings, some schools cheated on the data they submitted to US News & World Report, and others spent school funds trying to improve their rankings instead of in other ways which would have improved students' experiences at the school.  

**EXTRA CREDIT 55) What are the 10 simple rules Zook et. al. Suggest to follow when doing big data analytics?**

2. Recognize that privacy is more than a binary value
