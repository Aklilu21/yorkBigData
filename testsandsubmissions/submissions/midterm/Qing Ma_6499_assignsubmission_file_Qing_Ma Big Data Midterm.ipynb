**5) Test whether the two lists are equal too each other and if not what elements are shared between the two lists.**<br/>
<t/>list1 = ['1',2,'3',4,5]<br/>
<t/>list2 = [1,2,3,4,'5',6]<br/>

```python
# The code assumes the list contains elements in order
# So, [1, 2] and [2, 1] are not the same list
# [1, 2, 2] and [2, 1, 1] have no shared elements
# otherwise we need to deal with sets insead of lists
list1 = ['1',2,'3',4,5]
list2 = [1,2,3,4,'5',6]
# set the max steps for checking the lists
if len(list1) >= len(list2):
    step = len(list2)
else:
    step = len(list1)
    
joint = [] # the shared elements will be put in this list
equal = True # assume these two lists are the same

# if two lists have different length, then there are NOT equal
if len(list1) != len(list2):
    equal = False
# check if they have common elements
for i in range(step):
    if list1[i] != list2[i]:
        equal = False
    else:
        joint.append(list1[i])
# print the resualt        
if equal:
    print "These two lists are equal"
else:
    print "These two lists are NOT equal"
if joint == []:
    print "No shared element"
else:
    print "Shared elements are ", joint
```

**20) Suppose you have the following loop:**<br>
>things = [1,2,3,5,6,7,8,9,10]<br>
>previous_thing = 0<br>
>for thing in things:<br>
>>new_thing = thing * previous_thing<br>
>>x = 3+previous_thing<br>
>>previous_thing = thing<br>

**At the end of the loop:**
>**What does new_thing equal?**<br>
>**What does x equal?**
    
    


```python
# This loop runs 9 times. At end of each loop, previous_thing = thing, then at the last loop thing == 10 and previous_thing == 9
# Then, at the end, new_thing == 9*10 == 90, and x == 3 + 9 == 12
# Below, let's implement this loop and print the resualt to varify the answer
things = [1,2,3,5,6,7,8,9,10]
previous_thing = 0
for thing in things:
    new_thing = thing * previous_thing
    x = 3 + previous_thing
    previous_thing = thing
print 'new_thing: ', new_thing
print 'x: ', x
```
