```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
sns.distplot(x, kde=True, rug=True);
```

```python
sns.distplot(x, bins=50, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=False);
```

```python
x = np.random.normal(0, 1, size=50)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-5, 5, 100)

kernels = []
for x_i in x:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x, color=".2", linewidth=3);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```

```python
mean, cov = [0, 5], [(2, .75), (.25, 3)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 10000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(7, 7))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=2, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=50, shade=True);
```

```python
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```
