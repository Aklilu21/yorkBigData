```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
k = np.random.normal(size=120)
sns.distplot(k);
```

```python
sns.distplot(k, kde=False, rug=True);
```

```python
sns.distplot(k, hist=False, rug=True);
```

```python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * k.std() * k.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for k_i in k:

    kernel = stats.norm(k_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="y")

sns.rugplot(k, color="4", linewidth=3);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```

```python
sns.kdeplot(k, shade=True);
```

```python
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label="bw: 0.2")
sns.kdeplot(x, bw=2, label="bw: 2")
plt.legend();
```

```python
sns.kdeplot(x, shade=True, cut=0)
sns.rugplot(x);
```

```python
x = np.random.gamma(6, size=200)
sns.distplot(k, kde=False, fit=stats.gamma);
```

```python
mean, cov = [0,1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```

```python
a, ab = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=60, shade=True);
```

```python

g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```
