**19) Write a brief definition of what Data Latency and the difference between "real-time" and "near-time" data.**

Data latency describes time perspective of data; how frequently it is updated and how much of a time delay exists between an actual event happening and the related data being updated to reflect it.  

Real time refers to low latency where there is a small amount of time between a real event and when the data is updated to reflect it.  It would be considered 'up to date' all the time.
Near-time refers to latency that is less than a day but more than an hour; some lag exists between when it was 'current' and when it is out of date.

Standard deviation tells you the amount of variation or dispersion of the data values.  Low standard deviation means the values are close to the mean of the dataset while a high standard deviation means data points are spread out over a wider range of values.
