**22) In the following loop:**<br> 
>x = 1<br>

>>if x < 20001: <br>
>>>x = x-1<br>

**What does x equal to make our sure_thing False? **
        

```python
x = 1
sure_thing = True
while sure_thing:

    x = x +100
    if x < 20001:

        x = x-1
    else:
        sure_thing = False
```

```python
# sure thing will be false when x=20099. on the second last iteration, x=20000 which is smaller than 20001, therefore sure_thing is
#still True. On the last iteration, x=x-1=19999 so the while loop keeps going to the next iteration to make 
#x=x+100=19999+100=20099 which is >20001 which makes sure_thing=False
#this can be confirmed by running the code and printing the final value of x
print x
```
