# part 2: data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display # for displaying multiple output from one cell
```

```python
gnacs_x = """id|postedTime|body|None|['twitter_entiteis:urls:url']|['None']|\
['actor:languages_list-items']|gnip:language:value|twitter_lang|\
[u'geo:coordinates_list-items']|geo:type|None|None|None|None|\
actor:utcOffset|Something|Something|Something|Something|Something|Something|Something|Something|Something|\
actor:displayName|actor:preferredUsername|actor:id|gnip:klout_score|\
actor:followersCount|actor:friendsCount|actor:listedCount|\
actor:statusesCount|Tweet|None|None|None"""
colnames = gnacs_x.split('|')

# prevent the automatic compression of wide dataframes (add scroll bar)
pd.set_option("display.max_columns", None)
dfTwitter = pd.read_csv('yorkBigData-master/assignments/data/twitter_sample.csv', sep='|', names=colnames)
display(dfTwitter.head())
print dfTwitter.columns
print dfTwitter.shape
```

```python
# drop columns with 'something'
for col in dfTwitter.columns:
    if "Something" in col:
        dfTwitter = dfTwitter.drop(unicode(col),axis=1)
print dfTwitter.columns
print dfTwitter.shape
```

### slicing & combining

```python
actors = []
for col in dfTwitter.columns:
    if "actor" in col:
        actors.append(col)
print actors
dfActor = dfTwitter[actors]
dfActor = dfActor[dfActor["['actor:languages_list-items']"] == "['en']"]
display(dfActor.head())
```

```python
def isNull(val):
    return val == None or val == 'None'
def toFloat(val):
    if isNull(val): return 0.0
    else: return float(val)
def secToHour(val):
    return toFloat(val) / 3600.0

pd.options.mode.chained_assignment = None # make it silent instead of giving warnings
dfActor['actor:utcOffset'] = dfActor['actor:utcOffset'].map(secToHour)
dfActor['status/follower'] = dfActor['actor:statusesCount'] / dfActor['actor:followersCount']
display(dfActor.head())
display(dfActor.loc[1:10,'actor:followersCount':'status/follower'])
display(dfActor.describe())
```

### grouping

```python
dfCount = dfActor[dfActor["actor:followersCount"] < 5000]

for col in dfCount.columns:
    if not "Count" in col and not "utc" in col and not "status/follower" in col:
        dfCount = dfCount.drop(col, 1)

display(dfCount.head())
display(dfCount.groupby("actor:utcOffset").size())
display(dfCount.groupby("actor:utcOffset").mean())
```

```python
'''
dfStats = dfCount.aggregate(np.max).to_frame().transpose()
dfStats = dfStats.append(dfCount.aggregate(np.min).to_frame().transpose())
dfStats = dfStats.append(dfCount.aggregate(np.mean).to_frame().transpose())
dfStats = dfStats.append(dfCount.aggregate(np.std).to_frame().transpose())
dfStats = dfStats.drop('actor:utcOffset',1)
display(dfStats)
'''
dfGroup = dfCount.groupby('actor:utcOffset')
dfMean = dfGroup.aggregate(np.mean)
display(dfMean)
dfStd = dfGroup.aggregate(np.std)
display(dfStd)
```

# Visualizing the distribution of a dataset

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display # for displaying multiple output from one cell

sns.set(color_codes=True)
np.random.seed(sum(map(ord, "distributions")))
```

```python
#Plotting univariate distributions
planets = sns.load_dataset("planets")
display(planets.head())

plt.figure()
sns.kdeplot(planets['mass'], shade=True)
```

```python
x = planets['distance']
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(0, 10000, 200)

kernels = []
for x_i in x:
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="y")
sns.rugplot(planets['distance'], color="g")
```

```python
#Plotting bivariate distributions
plt.figure()
sns.jointplot(x="actor:statusesCount", y="actor:followersCount", data=dfStatusFollower);

plt.figure()
g = sns.jointplot(x="actor:statusesCount", y="actor:followersCount", data=dfStatusFollower, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="*")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

```python
#Visualizing pairwise relationships in a dataset
collision = sns.load_dataset("car_crashes")
display(collision.head())

g = sns.PairGrid(collision[['total','speeding','alcohol']])
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=10);
```

# Visualizing linear relationships

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display # for displaying multiple output from one cell

sns.set(color_codes=True)
np.random.seed(sum(map(ord, "regression")))
```

```python
collision = sns.load_dataset("car_crashes")
display(collision.head())

plt.figure()
g = sns.pairplot(x_vars=["speeding","alcohol","not_distracted","no_previous"], y_vars=["total"],
                 kind="reg", data=collision);
g.fig.set_size_inches(12,6)
```

```python
def formatTime(val):
    return float(val[0:-4])

pd.options.mode.chained_assignment = None # make it silent instead of giving warnings
exercise = sns.load_dataset("exercise")
exercise['fTime'] = exercise['time'].map(formatTime).astype('float')
display(exercise.head())
display(exercise.dtypes)

g = sns.lmplot(x="pulse", y="fTime", hue="kind", data=exercise)
g.fig.set_size_inches(15,6)
```

# Plotting with categorical data

```python
# import library and load data
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display # for displaying multiple output from one cell

sns.set(style="whitegrid", color_codes=True)
np.random.seed(sum(map(ord, "categorical")))
```

```python
# Categorical scatterplots
flights = sns.load_dataset("flights")
display(flights.head())

plt.figure()
plt.subplots(figsize=(12,4))
sns.swarmplot(x="month", y="passengers", hue="year", size=10, alpha=0.75, data=flights);
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
# Distributions of observations within categories
fmri = sns.load_dataset("fmri")
display(fmri.head())

plt.figure()
plt.subplots(figsize=(12,4))
sns.violinplot(x="signal", y="event", hue="region", palette="hls", split=True, data=fmri);
```

```python
plt.figure()
plt.subplots(figsize=(12,8))
sns.boxplot(x="subject", y="signal", hue="region", palette="husl", data=fmri.query("signal < 0.2"));
```
