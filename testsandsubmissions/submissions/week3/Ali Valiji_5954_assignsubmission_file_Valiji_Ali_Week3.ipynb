```python
df2 = df1[["gnip:klout_score","actor:followersCount", "actor:friendsCount", "actor:listedCount"]]

df2.head()
```

```python
df2.dtypes
```

```python
def floatify(val):
    if val == None or val == 'None':
        return 0.0
    else:
        return float(val)
```

```python
df2['gnip:klout_score'] = df2['gnip:klout_score'].map(floatify)


df2.dtypes
```

```python
df2 = df2.astype(float)

print df2.dtypes
```

```python
df2['fol/lc'] = df2['gnip:klout_score'] / df2['actor:listedCount']

print df2.head()
```

```python
pop_df = df1[df1["actor:followersCount"] >= 100]
```

```python
pop_df['gnip:klout_score'] = pop_df['gnip:klout_score'].map(floatify)
pop_df.head()
```

```python
pop_df.groupby("actor:followersCount").mean()
```

```python
pop_df.groupby("actor:statusesCount").size()
```

```python
lang_gb = pop_df[['twitter_lang',\
             'gnip:klout_score',\
             'actor:followersCount',\
             'actor:friendsCount',\
             'actor:statusesCount']].groupby('twitter_lang')

print lang_gb.head(4)
```

```python
lang_gb_mean = lang_gb.aggregate(np.mean)  

lang_gb_mean.head()
```
