```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x, kde=False, rug=True);
```

```python
sns.distplot(x, bins=25, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=True);
```

```python
x = np.random.normal(0, 1, size=25)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 150)

kernels = []
for x_i in x:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x, color=".2", linewidth=4);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1200).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

```python
f, ax = plt.subplots(figsize=(8, 8))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="r", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.x, df.y, cmap=cmap, n_levels=80, shade=True);
```
