```python
df2 = df1[["gnip:klout_score","actor:followersCount", "actor:friendsCount", "actor:listedCount"]]

df2.head(3)
```

```python
df2.dtypes
```

```python
def floatify(val):
    if val == None or val == 'None':
        return 0.0
    else:
        return float(val)
```

```python
df2['gnip:klout_score'] = df2['gnip:klout_score'].map(floatify)
df2.dtypes
```

```python
df2 = df2.astype(float)
df2.dtypes
```

```python
df2['fol/fr'] = df2['gnip:klout_score'] / df2['actor:followersCount']

df2.tail(3)
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "distributions")))
```

```python
y = np.random.normal(size=500)
sns.distplot(y, bins=40, kde=True, rug=True);
```

```python
sy = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * y.std() * y.size ** (-1 / 5.)
support = np.linspace(-4, 4, 100)

kernels = []
for y_i in y:

    kernel = stats.norm(y_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="g")

sns.rugplot(y, color=".2", linewidth=2);
```

```python
sns.kdeplot(y)
sns.kdeplot(y, bw=.2, label="bw: 0.2")
sns.kdeplot(y, bw=2, label="bw: 2")
plt.legend();
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 100)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 500).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="r");
```

```python
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="b")
g.plot_joint(plt.scatter, c="w", s=50, linewidth=1, marker="*")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "regression")))
tips = sns.load_dataset("tips")
```

```python
tips["small_tip"] = (tips.tip / tips.total_bill) < .10
sns.lmplot(x="total_bill", y="small_tip", data=tips,
           y_jitter=.02);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["+", "o"], palette="Set2");
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color='r');
```
