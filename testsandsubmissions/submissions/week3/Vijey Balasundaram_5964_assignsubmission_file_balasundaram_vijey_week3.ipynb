### Visualizing the distribution of a dataset

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

### Plotting univariate distributions

```python
x1 = np.random.normal(size=100)
sns.distplot(x1);
```

```python
sns.distplot(x1, kde=False, rug=True);
```

```python
sns.distplot(x1, bins=20, kde=False, rug=True);
```

```python
sns.distplot(x1, hist=False, rug=True);
```

```python
x2 = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x2.std() * x2.size ** (-1 / 5.) # ** operator = exponent
support = np.linspace(-4, 4, 200)

kernels = []
for x_i in x2:

    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(x2, color=".2", linewidth=3);
```

```python
density = np.sum(kernels, axis=0)
density /= integrate.trapz(density, support)
plt.plot(support, density);
```

```python
sns.kdeplot(x2, shade=True);
```

```python
sns.kdeplot(x2)
sns.kdeplot(x2, bw=.2, label="bw: 0.2")
sns.kdeplot(x2, bw=2, label="bw: 2")
plt.legend();
```

```python
sns.kdeplot(x2, shade=True, cut=0)
sns.rugplot(x2);
```

```python
x3 = np.random.gamma(6, size=200) #Fitting parametric distributions
sns.distplot(x3, kde=False, fit=stats.gamma);
```

### plotting bivariate distributions

```python
mean_1, cov_1 = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean_1, cov_1, 200)
df10 = pd.DataFrame(data, columns=["x", "y"])

```

```python
sns.jointplot(x="x", y="y", data=df10);
```

```python
z, a = np.random.multivariate_normal(mean_1, cov_1, 1500).T
with sns.axes_style("white"):
    sns.jointplot(x=z, y=a, kind="hex", color="k");

```

```python
sns.jointplot(x="x", y="y", data=df10, kind="kde");
```

```python
f1, ax1 = plt.subplots(figsize=(6, 6))
sns.kdeplot(df10.x, df10.y, ax=ax1)
sns.rugplot(df10.x, color="g", ax=ax1)
sns.rugplot(df10.y, vertical=True, ax=ax1);
```

```python
f2, ax2 = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df10.x, df10.y, cmap=cmap, n_levels=60, shade=True);
```

### jointplot function 

```python
g = sns.jointplot(x="x", y="y", data=df10, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```
