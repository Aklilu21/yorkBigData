```python
z = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * z.std() * z.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for z_i in z:

    kernel = stats.norm(z_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")

sns.rugplot(z, color=".2", linewidth=3);
```

```python
dens = np.sum(kernels, axis=0)
dens /= integrate.trapz(dens, support)
plt.plot(support, dens);
```
