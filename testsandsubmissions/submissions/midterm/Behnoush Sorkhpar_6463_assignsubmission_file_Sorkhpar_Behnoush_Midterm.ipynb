**EXTRA CREDIT 51) Why is Ordinary Least Squares Regression, called Ordinary Least Squares?**

Ordinary Least Squares describes a linear regression method for estimating the unknown parameters with the goal of minimizing error between estimated function and observations. It is called so since it was the first statistical procedure developed to fit a function to the data and it is arguably one of the simplest methods developed. It works based on 6 assumptions: model is linear in parameters; there is random sampling of observations; conditional mean should be zero; there is no multi-collinearity; there is homoscedasticity and no autocorrelation; and error terms are normally distributed.

**EXTRA CREDIT 52) What makes "big data" different from "small data"? (Be brief - 1 to 2 paragaphs MAX)**

Big data differs from small data for a number of characteristics that are unique to it. Big data is usually characterized by 5V: Volume, Velocity, Veracity, Value and Variety. The most obvious characteristic of big data that make it different from regular (small) data is the volume. Whereas small data comes in a volume and format that make it easily accessed, analyzed and actioned, big data comes in large volumes that is not easily manageable. Velocity refers to the speed at which data is generated. Whereas in small data, there is a dataset thats available and analyzed, in big data, new information is generated every second that makes handling of it more challenging. Veracity refers to quality of data  how trustworthy data is. With the volume and speed at which big data is generated, data quality becomes less controllable. This is another aspect that small data does not concern itself with as the amount of data make it easy to ensure quality of data is met. Variety refers to the different formats in which big data becomes available. Small data comes in limited digestible formats; however, big data can come in variety of formats that are not structured and therefore easily put into tables/charts. An example of this is pictures. Finally, value is the final difference between big data and small data. Small data has definite and specific attributes that help analyze the current situation. However, big data becomes useful when it adds value to not only understand current situation but also be able to build predictive models (and other types of models) based on the data.

**EXTRA CREDIT 54) Why do vicious feedback loops matter in data analytics?**

Vicious feedback loops are used to gain insight into the relevance of outputs of a model. When the outcome of a model is fed back into the model, it retrains itself and learns from the experience to generate a better, more realistic outcome. Feedback loops matter in big data as they help increase the accuracy of the model and the analysis of big data and subsequently result in better decision making based on the data.
