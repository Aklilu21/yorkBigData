Q1. For Q-1, you need to put the data file ‘Medicare_Hospital_Spending_by_Claim.csv’ at the same place as your notebook file.

Q1-a. You are given the Medicare spending dataset. The first thing to do is to remove 2 unused columns from the DataFrame in the following code segment. Provide your code as the answer.

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
claims = pd.read_csv('./Medicare_Hospital_Spending_by_Claim.csv', parse_dates=True) #, index_col=0)
claims.info()
print claims.columns
# Q1-a drop unused columns 'Measure Start Date' and 'Measure End Date'
claims.drop(['Measure Start Date', 'Measure End Date'], axis=1, inplace=True)
```

Q1-b Replace names
Continuing with the above code, we want to replace the original long column names with shorter ones.  Provide your code and the resulting new column names as the answer.
newColumns = ['Hospital', 'Provider ID', 'State', 'Period', 'Claim Type', 'Avg Spending Hospital', 'Avg Spending State', 'Avg Spending Nation', 'Percent Spending Hospital', 'Percent Spending State', 'Percent Spending Nation']

```python
# rename columns. add your code below
newColumns = ['Hospital', 'Provider ID', 'State', 'Period', 'Claim Type', 
             'Avg Spending Hospital', 'Avg Spending State', 'Avg Spending Nation',
             'Percent Spending Hospital', 'Percent Spending State', 'Percent Spending Nation']
print claims.columns
# rename columns
claims.columns = newColumns
print claims.columns
print claims[:10]
print claims['Avg Spending Hospital']
```

Q1-c) Continuing with the above code, we would like to investigate the statistics of average and percentage spending for different states, claim types, and hospitals.  In the code segment below, we print out the last 15 items of two columns for testing.  Do you see any issue in the data for statistics calculation?  Provide your comments as the answer. 

```python
# Ans: With the string characters present in the data, its not possible to perform Statistics calculation on them.
claims[['Percent Spending Hospital', 'Percent Spending State','Percent Spending Nation',
        'Avg Spending Hospital', 'Avg Spending State',  'Avg Spending Nation']].tail(15)
```

Q1-d) Continuing with the above code, we want to use the code below to enable statistic calculation for the DataFrame.  Provide your code and the resulting new print out as the answer.

```python
# convert %
origin1 = '%'
claims[['Percent Spending Hospital', 'Percent Spending State',
        'Percent Spending Nation']] = claims[['Percent Spending Hospital', 'Percent Spending State',
        'Percent Spending Nation']].replace(origin1,'',regex=True).astype('float')/100
# convert $ 
origin2 = '\$' # '[\$,]'
claims[['Avg Spending Hospital', 'Avg Spending State',
        'Avg Spending Nation']] = claims[['Avg Spending Hospital', 'Avg Spending State',
        'Avg Spending Nation']].replace(origin2,'' ,regex=True).astype('float')
claims[['Percent Spending Hospital', 'Avg Spending Hospital']].tail()
```

Q1-e) Continuing with the above code, we want to use the code segment below to extract only the TX state claims to study.  Provide your code and the resulting new print out as the answer.

```python
# extract TX only
tx_claims = claims[claims['State'] == 'TX']
tx_claims.tail()
```

Q1-f) Continuing from above, we want to study the average spending of different TX hospitals for different claim types.  Provide your code and the resulting new print out as the answer. 

```python
grouped_claims = tx_claims.groupby(['Hospital', 'Claim Type'])
# functions = ['min', 'mean', 'max']
# agg(functions)
results = grouped_claims[['Avg Spending Hospital', 'Avg Spending State', 'Avg Spending Nation']].mean()
# group avg spending
print results[['Avg Spending Hospital', 'Avg Spending State', 'Avg Spending Nation']].tail()
```

Q1-g) Continuing from above, we want to study the mean and max percentage spending of different TX hospitals for different claim types.  Provide your code and the resulting new print out as the answer. 

```python
# Define aggregation functions.  Add your code below
grouped_claims = tx_claims.groupby(['Hospital', 'Claim Type'])
functions = ['min', 'mean', 'max']
results = grouped_claims[['Avg Spending Hospital', 'Avg Spending State',
                        'Avg Spending Nation']].agg(functions)
print results[:7]
```

Q2-f. Continuing with the above code, add necessary Python code to create the DataFrame of previous monthly stock closing prices for the whole dataset. Provide your code as the answer.

```python
# create previous monthly closing stock prices.  Add your code below
close_px_m_prev = close_px_m.shift(1)
close_px_m_prev
```

Q2-g. Continuing with the above code, add necessary Python code to calculate the percentage of monthly and weekly change. Provide your code and the plots as the answer.

The percentage change series d(n) is defined as:

d(n) = (x(n) – x(n-1)) / x(n-1), where x(n) is the original Time Series.


```python

import matplotlib.pyplot as plt
plt.rc('figure', figsize=(10, 8))
%matplotlib inline
# Add your code for monthly percentage below
change_px_m = close_px_m/close_px_m_prev - 1
change_px_m.dropna().plot(subplots=True)
```

```python
# add your code for weekly percentage below

close_px_w = close_px.resample('W').mean()
change_px_w = close_px_w/close_px_w.shift(1) - 1
change_px_w.dropna().plot(subplots=True)
```
