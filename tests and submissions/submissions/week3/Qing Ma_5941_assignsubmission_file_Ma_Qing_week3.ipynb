#### Kernel density estimation

```python
sns.distplot(bell, hist = False, rug = True)
```

```python
bell = np.random.normal(0, 2, size = 40)
bandwidth = 1.06 * bell.std() * bell.size ** (-1 / 5.)
support = np.linspace(-8, 8, 200)
kernels = []
for bell_i in bell:
    kernel = stats.norm(bell_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color = 'b')
```

```python
density = np.sum(kernels, axis = 0)
density /= integrate.trapz(density, support)
plt.plot(support, density)
```

```python
sns.kdeplot(bell, shade = True)
```

```python
sns.kdeplot(bell, label = "default")
sns.kdeplot(bell, bw = .5, label = "bw: .1")
sns.kdeplot(bell, bw = 3, label = "bw: 2")
plt.legend()
```

```python
sns.kdeplot(bell, shade = True, cut = -1)
sns.rugplot(bell)
```

### Plotting bivariate distributions

```python
mean = [2, 4]
cov = [(1, .8), (.8, 1)]
mx = np.random.multivariate_normal(mean, cov, 1000)
df = pd.DataFrame(mx, columns = ['A', 'B'])
```

#### Scatterplots

```python
sns.jointplot(x = 'A',
              y = 'B', 
              data = df)
```

#### Hexbin plots

```python
a, b = mx.T
with sns.axes_style('white'):
    sns.jointplot(x = a, 
                  y = b, 
                  kind = 'hex', 
                  color = 'r')
```

#### Kernel density estimation

```python
sns.jointplot(x = b,
              y = a,
              data = df,
              kind = 'kde')
```

```python
fig, ax = plt.subplots(figsize = (7, 7))
sns.kdeplot(df.A, df.B, ax = ax)
sns.rugplot(df.A, color = 'b', ax = ax)
sns.rugplot(df.B, vertical = True, color = 'r', ax = ax)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(df.A, df.B, cmap=cmap, n_levels=50, shade=True);
```

```python
g = sns.jointplot(x = a[:200], y = b[:200], data = df, kind="kde", color="b")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="^")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$A$", "$B$");
```

## Statistical estimation within categories

### Bar plots

```python
sns.barplot(x='alone', y='survived', hue='sex', data=titanic)
```

```python
ages = []
for x in titanic.age:
    if x>0 and x<10:
        ages.append("10")
    elif x>=10 and x<20:
        ages.append('20')
    elif x>=20 and x<30:
        ages.append('30')
    elif x>=30 and x<40:
        ages.append('40')
    elif x>=40 and x<50:
        ages.append('50')
    elif x>=50 and x<60:
        ages.append('60')
    elif x>=60 and x<70:
        ages.append('70')
    elif x>=70 and x<80:
        ages.append('80')
    elif x>=80 and x<90:
        ages.append('90')
    else:
        ages.append('n/a')
```

```python
titanic["ages"] = ages
```

```python
sns.countplot(x='ages', data=titanic, palette='Greens_d')
```

```python
sns.countplot(y='ages', hue="class", data=titanic, palette='Blues_d')
```
