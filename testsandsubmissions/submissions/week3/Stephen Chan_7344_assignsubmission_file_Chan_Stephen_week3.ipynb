```python
m = np.random.normal(1, 2, size=20)
bandwidth = 1.5 * m.std() * m.size ** (-2 / 7.)
support = np.linspace(-5, 5, 300)

kernels = []
for m_i in m:

    kernel = stats.norm(m_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="g")

sb.rugplot(m, color=".3", linewidth=1);
```

```python
d = np.sum(kernels, axis=0)
d /= integrate.trapz(d, support)
plt.plot(support, d);
```
