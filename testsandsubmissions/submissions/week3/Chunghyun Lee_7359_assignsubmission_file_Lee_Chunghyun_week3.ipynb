# Visualizing the distribution of a dataset

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, 'distributions'))) 
```

### Plotting Univariate Dist. 

```python
a = np.random.normal(size=100)
sns.distplot(a);
```

```python
sns.distplot(a, kde=False, rug=True);
```

```python
sns.distplot(a, bins=20, kde=False, rug=True);
```

```python
sns.distplot(a, hist=False, rug=True);
```

```python
b = np.random.normal(0, 1, size=30)
band_w = 1.06 * b.std() * b.size ** (-1/5.)
spt= np.linspace(-4,4, 200)

krls = []
for b_i in b:
    
    krl = stats.norm(b_i, band_w).pdf(spt)
    krls.append(krl)
    plt.plot(spt, krl, color="g")
    
sns.rugplot(b, color=".2", linewidth=3);
```

```python
dn = np.sum(krls, axis=0)
dn /= integrate.trapz(dn, spt)
plt.plot(spt, dn);
```

```python
sns.kdeplot(b, shade=True);
```

```python
sns.kdeplot(b)
sns.kdeplot(b, bw=.2, label="bw: 0.2")
sns.kdeplot(b, bw=2, label="bw: 2")
plt.legend(); 
```

```python
sns.kdeplot(b, shade=True, cut=0)
sns.rugplot(b);
```

```python
p = np.random.gamma(6, size=200)
sns.distplot(p, kde=False, fit=stats.gamma);
```

### Plotting bivariate distributions

```python
mn, cv = [0, 1], [(1, .5), (.5, 1)]
dat = np.random.multivariate_normal(mn, cv, 200)
dframe = pd.DataFrame(dat, columns=["x", "y"])
```

```python
#Scatterplot
sns.jointplot(x="x", y="y", data=dframe);
```

```python
#Hexbin plots

x, y = np.random.multivariate_normal(mn, cv, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=y, y=y, kind="hex", color="k");
```

```python
#Kernel density estimation

sns.jointplot(x="x", y="y", data=dframe, kind='kde');
```

```python
#2-D KD Plot

f, ax = plt.subplots(figsize=(6,6))
sns.kdeplot(dframe.x, dframe.y, ax=ax)
sns.rugplot(dframe.x, color="g", ax=ax)
sns.rugplot(dframe.y, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(6,6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(dframe.x, dframe.y, cmap=cmap, n_levels=60, shade=True);
```

```python
# joint plot

G2 = sns.jointplot(x='x', y='y', data=dframe, kind='kde', color='m')
G2.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker='+')
G2.ax_joint.collections[0].set_alpha(0)
G2.set_axis_labels('$X$', '$Y$');
```

### Fitting different kinds of models

```python
anscombe = sns.load_dataset('anscombe')
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query("dataset == 'I'"),
          ci=None, scatter_kws={"s": 80});
```

```python
#polynominal regression model
sns.lmplot(x='x', y='y', data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={'s':80});
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={'s': 80});
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query("dataset =='III'"), # " to embedded ' (Note to self)
           robust=True, ci=None, scatter_kws={'s': 80});
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
          y_jitter=.03);
```

```python
#logistic regression line
sns.lmplot(x='total_bill', y='big_tip', data=tips,
          logistic=True, y_jitter=.03);
```

```python
#Lowess smoother #fit a non parametric regression
sns.lmplot(x='total_bill',y='tip',data=tips,lowess=True);
```

```python
#residplot 
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

```python
#If there is structure in the res, linear regression my not be appropriate
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
              scatter_kws={"s": 80});
```
