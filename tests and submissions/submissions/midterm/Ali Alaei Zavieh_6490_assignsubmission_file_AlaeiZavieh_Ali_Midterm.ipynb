**47) Give a very general and short definition of what an Analytical Model is or is supposed to do?**

```python
Models are representations of reality, they are based on simplifications and assumptions. Analytical models are a combination of rules which transform some inputs in different names and categories like explanatory variables, independent variables, predictors, decision variables, features and measure or metric into the outcome which is dependent variable to the model and inputs. 
```

**48) Briefly describe the differences between "supervised" and "unsupervised" learning.**

```python
Supervised learning is the task of inferring a function from labeled training data which training data consist of a set of training examples. Each example consists of an input object and a desired output value. Unsupervised Learning is the task of inferring a function to describe hidden structure from unlabeled data (a classification or categorization is not included in the observations).
```

**49) Briefly describe the differences between "classification" and "clustering".**

```python
Classification is a supervised learning technique where a training set and correctly defined observations are available. Clustering is a method of grouping objects in such a way that objects with similar features come together, and objects with dissimilar features go apart.
The key difference between clustering and classification is that clustering is an unsupervised learning technique used to group similar instances on the basis of features whereas classification is a supervised learning technique used to assign predefined tags to instances on the basis of features.
```

**50) Name 3 different "kinds" of Analytical Models and provide a brief definition of each. **

```python
Deterministic Equation: In deterministic models, the output of the model is fully determined by the parameter values and the initial conditions initial condition

Stochastic Equation: Stochastic models possess some inherent randomness. The same set of parameter values and initial conditions will lead to an ensemble of different outputs.

Hybrid: The Hybrid model is a "mixture" of both Deterministic and Stochastic. Its treatment is quite similar to the stochastic model. The presence of a single random variable in the model necessitates the consideration of the stochastic treatment.
```

```python
1. Acknowledge that data are people and can do harm.
2. Recognize that privacy is more than a binary value.
3. Guard against the re-identification of your data.
4. Practice ethical data sharing.
5. Consider the strengths and limitations of your data; big does not automatically mean better.
6. Debate the tough, ethical choices.
7. Develop a code of conduct for your organization, research community, or industry.
8. Design your data and systems for auditability.
9. Engage with the broader consequences of data and analysis practices.
10. Know when to break these rules.
```
