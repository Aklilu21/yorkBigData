**11) Calculate the descriptive statistice for each of the above central measures in Python on this list of numbers and print out the values:**<br/> 

[12,44,23,2,1,66,7,85,3,34,23,6,34,23,7,21,14,34,23,88]
<br/> 
*(Hint: it might be easier to import a function from some other library)*



```python
numberlist = pd.Series([12,44,23,2,1,66,7,85,3,34,23,6,34,23,7,21,14,34,23,14,34,23,88])
import numpy as np
import scipy.stats

k = numberlist.aggregate(np.mean)
median = numberlist.aggregate(np.median)
mode = scipy.stats.mode(numberlist)[0][0]

print  "mean %i" % k
print "median %i" % median 
print "mode %i" %  scipy.stats.mode(numberlist)[0][0]
```

**19) Write a brief definition of what Data Latency and the difference between "real-time" and "near-time" data.**

Data Latency refers to the timeframe between when the data is generated and when the end user is able to analyze the data. Real Time data is available for analysis very shortly the data collection process is initiated. Near-Time is defined as a situation where the data is ready for analysis >1 hr but less than a day after being initiated.
